| Mode | Description                  | Real-World Example                       |

 ## Error Handling

Handling Errors Like a professional


```python
try:
    age = int(input('Enter your age'))
    sub = 2026 - age
    print(sub)
except ValueError:
    print('Enter a digit')
```

```python
try:
    with open('nonexistent.txt', 'r') as file:
        content = file.read()
        print(content)
except FileNotFoundError:
    print('The file you are looking for doesnt exist')
```

```python
try:
    with open('nonexistent.txt', 'r') as file:
        content = file.read()
        print(content)
except Exception as e:
    print(e)
```

## Common Error

| Error Type        | What Causes It               | Example           |
|-------------------|------------------------------|-------------------|
| FileNotFoundError | File does not exist          | Wrong Filename    |
| PermissionError   | No permission to access file | Restricted folder |
| UnicodeError      | Wrong file encoding          | Corrupted text    |
| IOError           | General file problem         | Disk issue        |

### Key Idea
Exception handling prevents program from crashing unexpectedly


```python
def read_student_file(filename):
    """
    Safely reads a student's assignment
    return the content, if successful, or an error message if something goes wrong
    """
    try:
        with open(filename, 'r', encoding='utf-8') as file:
            content = file.read()
            print(f'Successfully read {filename}')
            return content
    except FileNotFoundError:
        print(f'{filename} was not found')

    except PermissionError:
        print(f'You have no permission to access {filename}')

    except UnicodeError:
        print(f'you cannot decode {filename}')

    except Exception as e:
        print(e)
        return None
```
