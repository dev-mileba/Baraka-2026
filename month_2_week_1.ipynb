## Functions

```
# Basic Function structure
def greet(): # defining the function
    print("Hello world")
    print("i love python")

greet() # Calling the function
```

## Functions with Parameters


```
def greet_person(name): # defining the function
    print(f"Hello, {name}! ")

greet_person("Michael")
greet_person("Daniel")
```

## Function with return statement

```
def add_numbers(a, b):
    return a + b

# result = add_numbers(2,3)
# print(result)

def square(number):
    return number ** 2

result = square(5)
print(result)
```

```
def get_stats(numbers):
    total = sum(numbers)
    avg = total / len(numbers)
    return total, avg

sum_nums, avg_nums = get_stats([10, 20, 30, 40, 50, 60, 70, 80, 90])
print(f"Total : {sum_nums}")
print(f"Average : {avg_nums}")
```

## Default Parameters


```
def greet_hi(name, greeting="Good morning"):
    return f"{greeting} {name}!"

greet_hi("Michael")
```

```
# Positional Arguments
def intro(name, age, city):
    print(f"My name is {name} and I am {age} years old. I live in {city}")

intro("Micheal", 18, "Coventry")
```

```
# arbitrary arguments
def sum_nums(*args):
    return sum(args)
sum_nums(1, 2, 3, 6)
```

## Functions with Loops and If Statement

```
# functions that count even numbers from a list of numbers
def count_evens(numbers):
    count = 0
    for num in numbers:
        if num % 2 == 0:
            count += 1
    return count

result_even = count_evens([1, 2, 3, 4, 5,6,8])
print(result_even)
```

## Part 4: Sets - Unique Collections

### What are Sets?

Sets are unordered collection of unique items

```
student_set = {'Kehinde', 'Rema', 'Vivian', 'Bright'}
type(student_set)
```

```
'Rema' in student_set
```

### Set Operations

```
fruit = {'apple', 'banana', 'orange'}
fruit.add('pear')
fruit.update(['mango', 'grape'])
fruit.remove('apple')
fruit.discard('pear')
print(fruit)
```

```
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}

union = set1 | set2 #or set1.union(set2)
intersection = set1 & set2
difference = set1 - set2 #what is in set1 and not set2
print(union)
print(intersection)
print(difference)
```
