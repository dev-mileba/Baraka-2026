### Hospital Management System

#### CRUD - Create, Read, Update, Delete

```python
import pandas as pd
```

```python
hospital_db = {
    'patients':[]
}
```

```python
def add_patient(name, age, gender, condition):
    patient = {
        'id':len(hospital_db['patients']) + 1,
        'name':name,
        'age':age,
        'gender':gender,
        'condition':condition
    }
    hospital_db['patients'].append(patient)
    print (f'{name} has been added successfully')
```

```python
add_patient('Jemima', 22, 'Female', 'Ulcer')
```

```python
def view_patient():
    if len(hospital_db['patients']) == 0:
        print('No patient records found')
        return
    df = pd.DataFrame(hospital_db['patients'])
    display(df) # It retains the normal data structure
```

```python
view_patient()
```

```python
# Access the list of dictionaries then target the one of interest
def update_patient(patient_id,key,new_value):
    for patient in hospital_db['patients']:
        if patient['id'] == patient_id:
            patient[key] = new_value
            print(f'Patient ID {patient_id} updated successfully')
            view_patient()
            return
    print('Patient not found')
```

```python
update_patient(2,'age',25 )
```

```python
view_patient()
```

```python
def delete_patient(patient_id):
    for patient in hospital_db['patients']:
        if patient['id'] == patient_id:
            hospital_db['patients'].remove(patient)
            print(f'Patient ID {patient_id} has been deleted')
            view_patient()
            return
    else:
        print('Patient not found')
```

```python
delete_patient(2)
```

```python
view_patient()
```

```python
# Main Menu
def main_menu():
    while True:
        print('\n============ Hospital Management System =============')
        print ('1 - Add Patient')
        print('2 - View Patient')
        print('3 - Update Patient')
        print('4 - Delete Patient')
        print('5 - Exit')

        choice = input('Enter your choice')
        if choice == '1':
            name = input('Enter Patient Name: ')
            age = int(input('Enter age: '))
            gender = input('Enter patient gender (M/F): ')
            condition = input('Enter patient condition: ')

            add_patient(name,age,gender,condition)

        elif choice == '2':
            view_patient()

        elif choice == '3':
            patient_id = int(input('Enter patient ID: '))
            key = input('Enter the key you wish to update: ')
            new_value = input(f'Enter the new value you wish to update {key}: ')

            update_patient(patient_id,key,new_value)

        elif choice == '4':
            patient_id = int(input('Enter patient ID: '))

            delete_patient(patient_id)

        elif choice == '5':
            print('Exiting the program...')
            break

        else:
            print('Invalid choice, Enter a number between 1 - 5')
```
