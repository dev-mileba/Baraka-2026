##  File Handling

```python
class_notes = [
    'Week 1: Intro to python - All learners participated\n',
    'Week 2: Data Types - 3 learners need extra help with dict\n',
    'Week 3: File Handling - Great engagement, practical example\n'
]
```

```python
with open('class_note.txt','w') as file:
    for note in class_notes:
        file.write(note)
print('Class notes saved successfully')
```

```python
print('Reading back from the class notes')
with open('class_note.txt', 'r') as file:
    content = file.read()
    print(content)
```

| Mode | Description                  | Real-World Example                       |
|------|------------------------------|------------------------------------------|
| 'r'  | Read only(default)           | Reading a recipe from a cookbook         |
| 'w'  | Write (overwrites exisiting) | Writing a new essay, replacing old draft |
| 'a'  | Append (adds to existing)    | Adding new entries to a diary            |
| 'r+  | Read and Write               | Editing a document you are reading       |
| 'rb' | Read Binary                  | Opening an image or video file           |
| 'wb' | Write Binary                 | Saving a photo you edited                |

 ## Error Handling

Handling Errors Like a professional


```python
try:
    age = int(input('Enter your age'))
    sub = 2026 - age
    print(sub)
except ValueError:
    print('Enter a digit')
```

```python
try:
    with open('nonexistent.txt', 'r') as file:
        content = file.read()
        print(content)
except FileNotFoundError:
    print('The file you are looking for doesnt exist')
```

```python
try:
    with open('nonexistent.txt', 'r') as file:
        content = file.read()
        print(content)
except Exception as e:
    print(e)
```
